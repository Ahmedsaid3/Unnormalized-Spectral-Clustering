```python
import osmnx as ox
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Graph construction 
def construct_graph(place_name):
    print(f"Constructing graph for: {place_name}")
    G = ox.graph_from_place(place_name, network_type='walk')
    return G
```

```python
place_name = "Istanbul Teknik Universitesi, Sariyer, Istanbul, Turkey"
G = construct_graph(place_name)
```

```python
ox.plot_graph(G)
```

```python
# Compute Adjacency Matrix and Degree Matrix
def compute_laplacian(G):
    A = nx.to_numpy_array(G)  # Adjacency matrix
    D = np.diag(A.sum(axis=1))  # Degree matrix
    L = D - A  # Unnormalized Laplacian
    
    # Display the matrices in a DataFrame for clarity
    print("Adjacency Matrix (A):")
    display(pd.DataFrame(A))
    
    print("Degree Matrix (D):")
    display(pd.DataFrame(D))
    
    print("Laplacian Matrix (L):")
    display(pd.DataFrame(L))

    return A, L
```

```python
A, L = compute_laplacian(G)
```

```python
# Inverse Power Method for Eigenvectors:
# Compute the first k eigenvectors of L using the inverse power method
# with optional spectral shift, orthogonalization, and deflation.
def inverse_power_method(L, shift=1e-4, num_vectors=6, max_iter=1000, tol=1e-6):

    n = L.shape[0]
    eigenvectors = []
    prev_vecs = []

    for _ in range(num_vectors):
        # Initialize a random vector v and normalize it
        v = np.random.rand(n)
        v = v / np.linalg.norm(v)

        for _ in range(max_iter):
            # Solve (L + shift*I) v_new = v   <-- Inverse iteration + shifting
            v_new = np.linalg.solve(L + shift * np.eye(n), v)   
            v_new = v_new / np.linalg.norm(v_new)

            # Orthogonalize v_new against all previously found eigenvectors
            for u in prev_vecs:  
                v_new = v_new - (u @ v_new) * u   # @ is a symbol for dot product
            v_new = v_new / np.linalg.norm(v_new)

            # Check for convergence: if change is below tolerance, stop iterating
            if np.linalg.norm(v_new - v) < tol:
                break
            v = v_new

        # Append the converged vector to our list
        eigenvectors.append(v)
        prev_vecs.append(v)
        
        # Estimate the corresponding eigenvalue via the Rayleigh quotient
        eigenvalue = np.dot(v.T, np.dot(L, v))
        print(eigenvalue)
        
        # Deflation: remove this eigen-component from L to find the next eigenvector
        L = L - np.outer(v, v) * eigenvalue
    
    # Stack the list of eigenvectors into an (n x num_vectors) array
    eigenvectors = np.array(eigenvectors).T
    
    # Display the eigenvectors
    print("\nEigenvectors from Inverse Power Method:")
    display(pd.DataFrame(eigenvectors))

    
    return eigenvectors
```

```python
# spectral method, eigenvectors produced by inverse power method is used
def spectral_kmeans(eigenvectors, n_clusters=6):
    kmeans = KMeans(n_clusters=n_clusters)
    labels = kmeans.fit_predict(eigenvectors)
    return labels
```

```python
labels_6 = spectral_kmeans(eigenvectors_6, n_clusters=6)
labels_9 = spectral_kmeans(eigenvectors_9, n_clusters=9)
labels_12 = spectral_kmeans(eigenvectors_12, n_clusters=12)
labels_15 = spectral_kmeans(eigenvectors_15, n_clusters=15)
```

```python
# direct KMeans on 2D coordinates 
def direct_KMeans(n_clusters=6):

    # Extract coordinates in node order
    x_pos = nx.get_node_attributes(G, 'x')
    y_pos = nx.get_node_attributes(G, 'y')
    coords = np.array([
        (x_pos[n], y_pos[n])
        for n in G.nodes()
        if n in x_pos and n in y_pos
    ])

    #  Apply K-Means
    labels_raw = KMeans(n_clusters=n_clusters, random_state=0).fit_predict(coords)
    return labels_raw
```

```python
labels_raw_6 = direct_KMeans(n_clusters=6)
labels_raw_9 = direct_KMeans(n_clusters=9)
labels_raw_12 = direct_KMeans(n_clusters=12)
labels_raw_15 = direct_KMeans(n_clusters=15)
```

### MODULARITY

```python
def calculate_modularity(A, labels):
    m = A.sum() / 2
    Q = 0.0
    for i in range(len(labels)):
        for j in range(len(labels)):
            if labels[i] == labels[j]:
                Q += (A[i, j] - (A.sum(axis=1)[i] * A.sum(axis=1)[j]) / (2 * m))
    return Q / (2 * m)
```

```python
# modularity results for spectral clustering labels (6, 9, 12, 15 cluster numbers)
modularity_6 = calculate_modularity(A, labels_6)
modularity_9 = calculate_modularity(A, labels_9)
modularity_12 = calculate_modularity(A, labels_12)
modularity_15 = calculate_modularity(A, labels_15)
print(f"Modularity of the clustering 6: {modularity_6}")
print(f"Modularity of the clustering 9: {modularity_9}")
print(f"Modularity of the clustering 12: {modularity_12}")
print(f"Modularity of the clustering 15: {modularity_15}")
```

```python
# modularity results for direct kmeans labels (6, 9, 12, 15 cluster numbers)
modularity_raw_6 = calculate_modularity(A, labels_raw_6)
modularity_raw_9 = calculate_modularity(A, labels_raw_9)
modularity_raw_12 = calculate_modularity(A, labels_raw_12)
modularity_raw_15 = calculate_modularity(A, labels_raw_15)
print(f"Modularity of the raw clustering 6: {modularity_raw_6}")
print(f"Modularity of the raw clustering 9: {modularity_raw_9}")
print(f"Modularity of the raw clustering 12: {modularity_raw_12}")
print(f"Modularity of the raw clustering 15: {modularity_raw_15}")
```

### VISUALIZATION

```python
def visualize_clusters(G_roadmap, labels, title):
    G_roadmap = G_roadmap.to_undirected() # for better visualization
    plt.figure(figsize=(10, 8))

    # Get 'x' and 'y' attributes for each node
    x_pos = nx.get_node_attributes(G_roadmap, 'x')
    y_pos = nx.get_node_attributes(G_roadmap, 'y')

    # Combine 'x' and 'y' into a single 'pos' dictionary
    pos = {
        node: (x_pos[node], y_pos[node])
        for node in G_roadmap.nodes()
        if node in x_pos and node in y_pos
    }

    # Draw the graph with the updated 'pos' layout
    nx.draw(
        G_roadmap,
        pos,
        node_color=labels,
        cmap=plt.cm.tab10,
        node_size=30,
        edge_color='gray',
        alpha=0.9
    )

    plt.title(title)
    plt.show()
```
